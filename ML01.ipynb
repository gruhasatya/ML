```python
from sklearn  import linear_model
import pandas as pd                          #Data Anaylsis and Mapulation library
import matplotlib.pyplot as plt              #visualization tool
import os
for dirname, _, files in os.walk('/train_and_test2.csv'):
  for filename in files:
    print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv('/train_and_test2.csv')   # To read file
data.head()                                  # file starting five will print
```

```python
data.info()                                  # Information about Data
```

```python
data.describe()                                 # Describing the data
```

```python
data.shape                                      # Data shape
```

```python
null_values = data.isnull().sum()               # Checking null values
null_values
```

```python
data = data.dropna(axis=0)                       # Dropping data
```

```python
data.describe()                                  # Checking
```

```python
train_data = data[0:900]                      # Training 80% data
train_data.shape                              # visualizing the shape
```

```python
test_data = data[900:]                       # Testing 20% data
test_data.shape                              # visualizing the shape
```

```python
features = ["Age", "Sex", "Embarked"]         # Feature Selection
train_in = train_data[features]               # Train input
train_out = train_data['2urvived']            # to be basic not changing the heading
test_in = test_data[features]
test_out = test_data['2urvived']
```

```python
linear = linear_model.LinearRegression()       # Model
linear.fit(train_in, train_out)
linear.score(train_in, train_out)
print('Coefficient: \n', linear.coef_)
print('Intercept: \n', linear.intercept_)
predicted = linear.predict(test_in)
```

```python
from sklearn.metrics import mean_squared_error  # just in case
er = mean_squared_error(test_out, predicted)
er
```
